# Section 2: Exploratory Data Analysis
There are a few descriptive statistics that can help provide an overview of the dataset. These statistics are the standard deviation and the mean. The standard deviation provides an overview of the causes of deaths, it will be able to show the differences between some of the countries; some countries may have a high amount of nutritional deficiency deaths but others don't and the standard deviation is able to reflect that based on how high or low these standard deviations are. The mean is also valuable because you can generally see what cause of death should be the biggest and most probable.

```python
df_no_code = df.drop(axis=1, columns="Code")
#Drops the "Code" column to use groupby feature on the country
```

```python
#Finding out the STD for each country's causes of deaths throughout the years
df_std = df_no_code.groupby("Country/Territory").agg("std").reset_index()
df_std
```

In this standard deviation dataset we can see that there are a variety of different standard deviations for each country. Some have STD that are huge and some not so much. For example, if I were to compare the U.S. with a developing country like Ethiopia, you would see that the STD for the causes of deaths would be vastly different. Higher STD usually means that the cause of death is very unpredictable or hard to treat, varying from year to year. The difference could be interpreted as the difference in development but also in the country's lifestyle.

```python
df_std[(df_std["Country/Territory"] == "United States") | (df_std["Country/Territory"] == "Ethiopia")]
```

Moving onto mean, the number for the mean is dependent on the country and you can somewhat tell the difference between two countries. Using the same example from before, if you compare the US to Ethiopia, you would see that the means for some of the causes are complete opposites of each other with one being huge and the other being small. This is representative of these two countries' difference in development. 

```python
#Finding out the mean for each country's causes of deaths throughout the years
df_mean = df_no_code.groupby("Country/Territory").agg("mean").reset_index()
df_US_Ethiopia =df_mean[(df_mean["Country/Territory"] == "United States") | (df_std["Country/Territory"] == "Ethiopia")]
```

# Section 4: Data Science


One question for me is to see if I can find out if the difference in development between countries is able to be uncovered within the causes of deaths. In other words, I am trying to answer the question, "Are causes of deaths a good indication of whether or not a country is developed or developing?" To do this, I will utilize kmeans clustering in order to discover patterns and cluster the countries based on whether or not the country may be developed or developing. I will have to take into account the total number of deaths and use that to convert all causes of deaths into a percentage of the total, then with that I can use the model to predict if the causes are related to the country development.

```python
#First I import KMeans and create the model
from sklearn.cluster import KMeans
kmeans = KMeans(2)
```

```python
#Sorry about this, I'm doing a whole bunch of the same thing to make each column.

df_countries = df_no_code.groupby("Country/Territory").agg("sum").reset_index()
df_countries["Total Deaths"] = (df_countries["Acute Hepatitis"] + df_countries["Alcohol Use Disorders"] 
+ df_countries["Alzheimer's Disease and Other Dementias"] + df_countries["Cardiovascular Diseases"] + df_countries["Chronic Kidney Disease"] 
+ df_countries["Chronic Respiratory Diseases"] + df_countries["Cirrhosis and Other Chronic Liver Diseases"] + df_countries["Conflict and Terrorism"] 
+ df_countries["Diabetes Mellitus"] + df_countries["Diarrheal Diseases"] + df_countries["Digestive Diseases"] + df_countries["Drowning"] 
+ df_countries["Drug Use Disorders"] + df_countries["Environmental Heat and Cold Exposure"] + df_countries["Exposure to Forces of Nature"] + df_countries["Fire, Heat, and Hot Substances"] 
+ df_countries["HIV/AIDS"] + df_countries["Interpersonal Violence"] + df_countries["Lower Respiratory Infections"] + df_countries["Malaria"]
+ df_countries["Maternal Disorders"] + df_countries["Meningitis"] + df_countries["Neonatal Disorders"] + df_countries["Neoplasms"] + df_countries["Nutritional Deficiencies"]
+ df_countries["Parkinson's Disease"] + df_countries["Poisonings"] + df_countries["Protein-Energy Malnutrition"] + df_countries["Road Injuries"] + df_countries["Self-harm"] + df_countries["Tuberculosis"])

#I'm making a percentage column for each cause of death
df_countries["Acute Hepatitis %"] = df_countries["Acute Hepatitis"] / df_countries["Total Deaths"]
df_countries["Alcohol Use Disorders %"] = df_countries["Alcohol Use Disorders"] / df_countries["Total Deaths"]
df_countries["Alzheimer's Disease and Other Dementias %"] = df_countries["Alzheimer's Disease and Other Dementias"] / df_countries["Total Deaths"]
df_countries["Cardiovascular Diseases %"] = df_countries["Cardiovascular Diseases"] / df_countries["Total Deaths"]
df_countries["Chronic Kidney Disease %"] = df_countries["Chronic Kidney Disease"] / df_countries["Total Deaths"]
df_countries["Chronic Respiratory Diseases %"] = df_countries["Chronic Respiratory Diseases"] / df_countries["Total Deaths"]
df_countries["Cirrhosis and Other Chronic Liver Diseases %"] = df_countries["Cirrhosis and Other Chronic Liver Diseases"] / df_countries["Total Deaths"]
df_countries["Conflict and Terrorism %"] = df_countries["Conflict and Terrorism"] / df_countries["Total Deaths"]
df_countries["Diabetes Mellitus %"] = df_countries["Diabetes Mellitus"] / df_countries["Total Deaths"]
df_countries["Diarrheal Diseases %"] = df_countries["Diarrheal Diseases"] / df_countries["Total Deaths"]
df_countries["Digestive Diseases %"] = df_countries["Digestive Diseases"] / df_countries["Total Deaths"]
df_countries["Drowning %"] = df_countries["Drowning"] / df_countries["Total Deaths"]
df_countries["Drug Use Disorders %"] = df_countries["Drug Use Disorders"] / df_countries["Total Deaths"]
df_countries["Environmental Heat and Cold Exposure %"] = df_countries["Environmental Heat and Cold Exposure"] / df_countries["Total Deaths"]
df_countries["Exposure to Forces of Nature %"] = df_countries["Exposure to Forces of Nature"] / df_countries["Total Deaths"]
df_countries["Fire, Heat, and Hot Substances %"] = df_countries["Fire, Heat, and Hot Substances"] / df_countries["Total Deaths"]
df_countries["HIV/AIDS %"] = df_countries["HIV/AIDS"] / df_countries["Total Deaths"]
df_countries["Interpersonal Violence %"] = df_countries["Interpersonal Violence"] / df_countries["Total Deaths"]
df_countries["Lower Respiratory Infections %"] = df_countries["Lower Respiratory Infections"] / df_countries["Total Deaths"]
df_countries["Malaria %"] = df_countries["Malaria"] / df_countries["Total Deaths"]
df_countries["Maternal Disorders %"] = df_countries["Maternal Disorders"] / df_countries["Total Deaths"]
df_countries["Meningitis %"] = df_countries["Meningitis"] / df_countries["Total Deaths"]
df_countries["Neonatal Disorders %"] = df_countries["Neonatal Disorders"] / df_countries["Total Deaths"]
df_countries["Neoplasms %"] = df_countries["Neoplasms"] / df_countries["Total Deaths"]
df_countries["Nutritional Deficiencies %"] = df_countries["Nutritional Deficiencies"] / df_countries["Total Deaths"]
df_countries["Parkinson's Disease %"] = df_countries["Parkinson's Disease"] / df_countries["Total Deaths"]
df_countries["Poisonings %"] = df_countries["Poisonings"] / df_countries["Total Deaths"]
df_countries["Protein-Energy Malnutrition %"] = df_countries["Protein-Energy Malnutrition"] / df_countries["Total Deaths"]
df_countries["Road Injuries %"] = df_countries["Road Injuries"] / df_countries["Total Deaths"]
df_countries["Self-harm %"] = df_countries["Self-harm"] / df_countries["Total Deaths"]
df_countries["Tuberculosis %"] = df_countries["Tuberculosis"] / df_countries["Total Deaths"]
df_countries

```

```python
#Second I gather up all the numerical categories into a variable
percentage_categories = ["Acute Hepatitis %","Alcohol Use Disorders %", 
"Alzheimer's Disease and Other Dementias %", "Cardiovascular Diseases %", "Chronic Kidney Disease %", 
"Chronic Respiratory Diseases %","Cirrhosis and Other Chronic Liver Diseases %","Conflict and Terrorism %", 
"Diabetes Mellitus %", "Diarrheal Diseases %", "Digestive Diseases %","Drowning %", 
"Drug Use Disorders %", "Environmental Heat and Cold Exposure %", "Exposure to Forces of Nature %", "Fire, Heat, and Hot Substances %", 
"HIV/AIDS %", "Interpersonal Violence %", "Lower Respiratory Infections %", "Malaria %",
"Maternal Disorders %", "Meningitis %", "Neonatal Disorders %", "Neoplasms %", "Nutritional Deficiencies %",
"Parkinson's Disease %", "Poisonings %","Protein-Energy Malnutrition %", "Road Injuries %", "Self-harm %", "Tuberculosis %"]
#Third I train the model using a dataframe that I created to contain the sum of all causes of deaths by country
model = kmeans.fit(df_countries[percentage_categories])

```

```python
#Last I use this model to predict and cluster the dataset and all the countries into two types of countries
df_countries["Developed/Developing"] = model.predict(df_countries[percentage_categories])
df_countries
```

```python
#Looking at United States and Ethiopia again
df_developed_or_not1 = df_countries[(df_countries["Country/Territory"] == "United States") | (df_countries["Country/Territory"] == "Ethiopia")]
df_developed_or_not1[["Country/Territory","Developed/Developing"]]
```

Seems like the cluster that the United States is in is developed countries and the one Ethiopia is in is developing countries and the model was successful in being able to predict whether or not a country is developed or not.
